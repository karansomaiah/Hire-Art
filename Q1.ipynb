```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_table('HireArtdata.csv', sep=',')
```

```python
day = []
month = []
year = []

for x in data['Date of Contact']:
    date_elements = x.split("/")
    day.append(int(date_elements[1]))
    month.append(int(date_elements[0]))
    year.append(int(date_elements[2]))
```

```python
# some pre-processing
data['Day'] = day
data['Month'] = month
data['Year'] = year
```

```python
data.head()
```

### Plot showing the number of Contacts each month for all the years 

```python
plt.plot(data['Month'].value_counts().sort_index())
plt.show()
```

```python
# Percentage of contacts in the month of october
data['Month'].value_counts().sort_index()[10]/sum(data['Month'].value_counts().sort_index().values)
```

```python
# function for getting yearly data
def yearly_contacts(year, data):
    temp = data[data['Year'] == year]
    series = temp['Month'].value_counts().sort_index()
    return(series)
```

***

Let's find out for the number of contacts for each month for year 2013

```python
data_2013 = yearly_contacts(2013, data)
```

```python
plt.plot(data_2013)
plt.show()
```

Since, we only have data for 3 monthts i.e. October, November and December, the plot is not significant. But, it is still evident there is a huge jump in the number of contacts in October compared to November and December.

Lets find out for year 2014

```python
data_2014 = yearly_contacts(2014, data)
plt.plot(data_2014)
plt.show()
```

The above plot is again indicative of October being the month with the highest number of contacts to the clients

Lets find out for year 2015

```python
data_2015 = yearly_contacts(2015, data)
plt.plot(data_2015)
plt.show()
```

Lets find out for year 2016

```python
data_2016 = yearly_contacts(2016, data)
plt.plot(data_2016)
plt.show()
```

Lets find out for year 2017

```python
data_2017 = yearly_contacts(2017, data)
plt.plot(data_2017)
plt.show()
```

**Conclusion**: Throughout all the years, we see a trend that has a peak in the month of October. Thus, it is a good time to schedule a product upgrade annocement. Also, just by looking at yearly plots it is pretty clear that this is seasonal data. You observe a certain trend towards the approaching of each month. There is a slight peak in June each year, followed by a boost in the numeber of contacts in the month of October. Throughout the years, majority of peaks are in October followed by September and June.
